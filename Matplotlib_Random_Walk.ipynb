```python
def rand_walk(step_num):
    walk = []
    step_choice = choice([1,-1])
    walk.append(step_choice)
    for i in range (1,step_num):
            step_choice_2 = choice ([1,-1])
            next_step = walk[i-1]+choice([1,-1])
            walk.append(next_step)
    return walk
```

```python
random_w = rand_walk(100)
#print(len(random_w))
```

```python
def plot_walk(walk):
    x = list(range(1, len(walk)+1))
    
    plt.plot(x,walk,'k--')
    plt.colorbar
    # plt.xlim(0,1.0)
    # plt.ylim(0,50)
    plt.title('This is a rnadom walk graph')
    plt.xlabel('Number of steps')
    plt.ylabel('Distance from the origin')
    plt.show()
```

```python
plot_walk(rand_walk(1000))
```

```python
def plot_walks(num,walk_steps):
    lab_list = list(range(1,num+1))
    for i in range(0,num):
        
        x = list(range(1, walk_steps+1))
        plt.plot(x,rand_walk(walk_steps), label = 'Plot Number' + ' '+str(lab_list[i]),)
        # plt.xlim(0,1.0)
        # plt.ylim(0,50)
        plt.title('Our random walk')
        plt.xlabel('Number of steps')
        plt.ylabel('Distance from the origin')
        plt.legend(loc='lower left')
    plt.show()
```

# Sierpisiki Triangle Solve - Creating a triangle

First transformation
    xn+1 = 0.5xn
    yn+1 = 0.5yn

Second Transformation 
    xn+1 = 0.5Xn+0.5
    yn=1 = 0.5yn+0.5

Third Transformation
    xn+1 = 0.5xn+1
    yn+1 = 0.5yn

```python
def trans_1(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x
    y1 = 0.5 * y
    return x1, y1
    
def trans_2(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 0.5
    y1 = 0.5 * y + 0.5
    return x1, y1
    
def trans_3(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 1
    y1 = 0.5 * y
    return x1, y1
    
transformations = [trans_1, trans_2, trans_3]
a1 = [0]
b1 = [0]
a,b = 0,0

for i in range(10000):
    trans = choice(transformations)
    a,b = trans((a,b))
    a1.append(a)
    b1.append(b)
```

```python
plt.rc('figure', figsize=(16,16))
```

```python
plt.plot(a1,b1,'o')
plt.savefig("Myfigure.png")
```
